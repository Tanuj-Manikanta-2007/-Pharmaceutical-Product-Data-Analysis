```python

Q1 = df['price(₹)'].quantile(0.25)
Q3 = df['price(₹)'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df = df[(df['price(₹)'] >= lower_limit) & (df['price(₹)'] <= upper_limit)]
df
```

```python
import re

def extract_strength(text):
    text = str(text)
    match = re.search(r'(\d+)', text)   # Extract first number
    if match:
        return int(match.group(1))      # Convert to number
    else:
        return None
```

```python
for df in des_df:
    df['strength'] = df['short_composition1'].apply(extract_strength)
```

```python
for i in range(len(des_df)):
    data = des_df[i]
    label = names[i]

    # Count most common strengths
    strength_counts = data['strength'].value_counts().head(10)

    # Plot pie chart
    plt.figure(figsize=(5,5))
    plt.pie(strength_counts.values, labels=strength_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(f"Top Strength Distribution in {label} Medicines")
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

df_corr = df_cp.copy()
le = LabelEncoder()

for col in ['manufacturer_name', 'pack_size_label', 'short_composition1', 'short_composition2']:
    df_corr[col] = le.fit_transform(df_corr[col].astype(str))
```

```python
corr_matrix = df_corr[['price(₹)',
                       'manufacturer_name',
                       'pack_size_label',
                       'short_composition1',
                       'short_composition2']].corr()

corr_matrix
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Between Price and Categorical Features")
plt.show()
```
